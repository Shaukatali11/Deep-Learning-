# Feature Engineering 

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import OneHotEncoder
```

# Step-1: We have Create a DataFrame 

```python
data = pd.DataFrame({
    'CustomerID': [1, 2, 3, 1, 2],
    'PurchaseDate': ['2022-01-01', '2022-02-15', '2022-03-10', '2022-03-20', '2022-04-05'],
    'ProductCategory': ['Electronics', 'Clothing', 'Books', 'Electronics', 'Books'],
    'ProductPrice': [100, 50, 30, 120, 25],
    'PurchaseQuantity': [2, 1, 3, 1, 2],
    'CustomerAge': [30, 25, 40, 30, 25],
    'CustomerRegion': ['North', 'South', 'East', 'North', 'West'],
    'PaymentMethod': ['Credit Card', 'PayPal', 'Credit Card', 'Credit Card', 'PayPal'],
    'TotalPurchaseAmount': [200, 50, 90, 120, 50],
    'HighValuePurchase': [1, 0, 0, 1, 0],
    'TotalPurchaseAmount_Avg': [160, 75, 90, 160, 75],
    'ProductCategory_Electronics': [1, 0, 0, 1, 0],
    'ProductCategory_Clothing': [0, 1, 0, 0, 0],
    'ProductCategory_Books': [0, 0, 1, 0, 1],
    'PaymentMethod_Credit Card': [1, 0, 1, 1, 0],
    'PaymentMethod_PayPal': [0, 1, 0, 0, 1]
})
```

# when we have create DataFrame Our 'PurchaseDate' is object we need to conert it into datetime formate

```python
# Convert PurchaseDate to datetime format
data['PurchaseDate'] = pd.to_datetime(data['PurchaseDate'])
```

```python
data
```

```python
data['PurchaseDate'].dtypes
```

# Apply Different Different to feature technique

```python
# Step 1: Extract Year and Month from PurchaseDate
data['Year'] = data['PurchaseDate'].dt.year
data['Month'] = data['PurchaseDate'].dt.month
```

```python
data
```

```python
# Step 2: Calculate Average Purchase Amount per Customer
avg_purchase_per_customer = data.groupby('CustomerID')['TotalPurchaseAmount'].mean()
data = data.merge(avg_purchase_per_customer, on='CustomerID', suffixes=('', '_Avg'))
```

```python
avg_purchase_per_customer
```

```python
data
```

```python
# Step 3: Encode ProductCategory and PaymentMethod using one-hot encoding
encoder = OneHotEncoder(sparse=False)
```

```python
data['ProductCategory'].nunique()
```

```python
data['PaymentMethod'].nunique()
```

```python
encoded_categories = encoder.fit_transform(data[['ProductCategory', 'PaymentMethod']])
```

```python
encoded_categories
```

```python
encoded_df = pd.DataFrame(encoded_categories, columns=encoder.get_feature_names(['ProductCategory', 'PaymentMethod']))
data = pd.concat([data, encoded_df], axis=1)
```

```python
encoded_df
```

```python
data
```

```python
# Step 4: Create Age Groups for Customers
bins = [0, 30, 40, 50, np.inf]
labels = ['Young', 'Middle-Aged', 'Late Middle-Aged', 'Senior']
```

```python
data['AgeGroup'] = pd.cut(data['CustomerAge'], bins=bins, labels=labels, right=False)
```

```python
data
```

```python
data['AgeGroup'] = data['CustomerAge'].apply(lambda age: 'Young' if age < 30 else ('Middle-Aged' if age < 50 else 'Senior'))
```

```python
data
```

```python
# Step 5: Calculate Purchase Frequency
purchase_frequency = data.groupby('CustomerID').size().reset_index(name='PurchaseFrequency')
```

```python
data = data.merge(purchase_frequency, on='CustomerID')
```

```python
data
```

```python
# Step 6: Normalize Numerical Features (ProductPrice, PurchaseQuantity, CustomerAge)
scaler = MinMaxScaler()
numerical_cols = ['ProductPrice', 'PurchaseQuantity', 'CustomerAge']
```

```python
data[numerical_cols] = scaler.fit_transform(data[numerical_cols])
```

```python
data
```

```python
# Step 7: Create Seasonality Features (Month)
data['IsHolidaySeason'] = data['Month'].isin([12, 1])  # Assume December and January as holiday season
```
